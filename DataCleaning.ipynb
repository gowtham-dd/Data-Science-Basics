```python
import pandas as pd
airbnb = pd.read_csv('https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/airbnb.csv?raw=true', index_col = 'Unnamed: 0')
display(airbnb)
```

```python
# airbnb.isnull().sum()
# airbnb["coordinates"]=airbnb["coordinates"].str.replace("(","")
# airbnb["coordinates"]=airbnb["coordinates"].str.replace(")","")
# # display(airbnb)
# a=airbnb["coordinates"].str.split(",",expand=True)
# display(a)
# airbnb["Lat"]=a[0]
# airbnb["Lon"]=a[1]
# # display(airbnb)
airbnb["listing_id"].nunique()#number of unique
airbnb["listing_id"].unique()
```

```python
#cleaning
airbnb["price"]=airbnb["price"].str.replace("$","")
airbnb["price"]=airbnb["price"].astype(float)#if we dont remove dollor we cant clean it based on the data type
airbnb.info()
```

```python
b=airbnb["listing_id"].duplicated(keep="first")
airbnb[b].sort_values("listing_id").shape
```

```python
b = airbnb["listing_id"].duplicated(keep="first")
c = airbnb[b].sort_values("listing_id")
display(c)
```

```python
airbnb["room_type"].unique()
airbnb["room_type"]=airbnb["room_type"].str.replace("Entire home/apt","private")
airbnb["room_type"]=airbnb["room_type"].str.replace("Private room","private")
airbnb["room_type"]=airbnb["room_type"].str.replace("Shared room","shared room")

airbnb
```

```python
airbnb.groupby("room_type").agg({"price":["mean","max","min"]})
```

```python
# Impute price based on conditions
airbnb.loc[(airbnb['price'].isna()) & (airbnb['room_type'] == 'home'), 'price'] = 208.7
airbnb.loc[(airbnb['price'].isna()) & (airbnb['room_type'] == 'private'), 'price'] = 89.02
airbnb.loc[(airbnb['price'].isna()) & (airbnb['room_type'] == 'shared room'), 'price'] = 72.41
```
