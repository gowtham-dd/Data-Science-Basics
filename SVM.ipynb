```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
iris.feature_names
```

```python
iris.target_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
df[df.target==2].head()
```

```python
df['Flower Name']=df.target.apply(lambda x:iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0=df[df.target==0]
df2=df[df.target==1]
df3=df[df.target==2]
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color="green",marker="+")
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color="blue",marker="*")
```

```python

plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color="green",marker="+")
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color="blue",marker="*")
```

```python
x=df.drop(['target','Flower Name'],axis='columns')
x.head()
```

```python
y=df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
import pandas as pd
from sklearn.datasets import load_digits
digit=load_digits()
digit.feature_names
digit.target_names
```

```python
df=pd.DataFrame(data=digit.data,columns=digit.feature_names)
df.head()
df['target']=digit.target
df.head()
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
df3=df[df.target==3]
df4=df[df.target==4]

plt.scatter(df0['pixel_0_1'],df0['pixel_0_2'],color="green",marker="+")
plt.scatter(df2['pixel_0_1'],df2['pixel_0_2'],color="blue",marker="*")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(['target'],axis='columns'),df.target,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
