```python
import matplotlib.pyplot as plt

# We'll sample 5 restaurants for a clean plot
sample_df = df[['name', 'rating', 'cost', 'rating_count']].dropna().head(5)

names = sample_df['name']
ratings = sample_df['rating']
costs = sample_df['cost']
rating_counts = sample_df['rating_count']

# Create subplot
fig, ax = plt.subplots(figsize=(12, 6))

# Plot ratings vs cost as scatter
ax.scatter(costs, ratings, color='teal', s=100)

# Annotate each point with restaurant name
for i in range(len(sample_df)):
    ax.text(costs.iloc[i] + 5, ratings.iloc[i], names.iloc[i], fontsize=9)

# Add vertical and horizontal lines
ax.axvline(x=300, color='gray', linestyle='--', label='Cost = 300')
ax.axhline(y=4.0, color='purple', linestyle='--', label='Rating = 4.0')

# Add shaded area (span) to highlight sweet spot
ax.axvspan(150, 300, facecolor='green', alpha=0.1, label='Affordable Range')
ax.axhspan(4.0, 5.0, facecolor='blue', alpha=0.1, label='High Rating Zone')

# Labels and title
ax.set_xlabel("Cost for One (₹)", fontsize=12)
ax.set_ylabel("Rating", fontsize=12)
ax.set_title("Swiggy Restaurants: Rating vs Cost", fontsize=14)
ax.set_xlim(0, max(costs) + 50)
ax.set_ylim(0, 5.5)
ax.legend(loc='lower right')

plt.tight_layout()
plt.show()
```

## Twin

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame (replace this with your actual cleaned Swiggy dataset)
# Make sure these columns exist in your DataFrame
# For demonstration, I'll use a small sample
data = {
    'name': ['A', 'B', 'C', 'D', 'E'],
    'cost': [100, 200, 300, 400, 500],
    'rating': [3.8, 4.2, 4.0, 4.5, 3.6],
    'rating_count': [120, 250, 80, 400, 30]
}
df = pd.DataFrame(data)

# Create figure and main axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot rating vs cost
ax1.plot(df['cost'], df['rating'], color='blue', marker='o', label='Rating')
ax1.set_xlabel('Cost for One (₹)')
ax1.set_ylabel('Rating', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_ylim(0, 5)

# Create secondary y-axis
ax2 = ax1.twinx()
ax2.plot(df['cost'], df['rating_count'], color='green', marker='s', linestyle='--', label='Rating Count')
ax2.set_ylabel('Rating Count', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Title and grid
plt.title("Twin Axis Plot: Rating and Rating Count vs Cost")
ax1.grid(True)

# Optional: Add legend
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

plt.tight_layout()
plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt

# Example bin ranges that suit the sample data (150 to 450)
bin_edges = [100, 200, 300, 400, 500]

fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(df['Avg_Cost'], bins=bin_edges, color='skyblue', edgecolor='black',histtype="step")

# Customizing the plot
ax.set_title("Histogram of Average Cost", fontsize=14)
ax.set_xlabel("Avg Cost (₹)", fontsize=12)
ax.set_ylabel("Number of Restaurants", fontsize=12)
ax.set_xticks(bin_edges)
ax.legend()
plt.tight_layout()
plt.show()
```

## Contour Plot

```python
# use only when to see z changes the input of x and y
#make x and y in mish grid using numpy
# 2 var as usual and z is represented by isolines
# x and y are independent z is dependent
xlist=np.linspace(-3,3,100)
ylist=np.linspace(-3,3,100)
x,y=np.meshgrid(xlist,ylist)
z=np.sin(np.sqrt(x**2+y**2))
fig,ax=plt.subplots()
con=ax.contourf(x,y,z)
fig.colorbar(con)
plt.show()
```

## Annotation

```python
import matplotlib.pyplot as plt

# Define bins appropriate for the data
bin_edges = [100, 200, 300, 400, 500]

fig, ax = plt.subplots(figsize=(8, 5))

# Plot histogram and capture bin counts and edges
counts, bins, patches = ax.hist(df['Avg_Cost'], bins=bin_edges, color='lightgreen', edgecolor='black')

# Annotate each bar with the count
for count, bin_edge in zip(counts, bins[:-1]):
    ax.text(bin_edge + 25, count + 0.1, int(count), ha='center', fontsize=11, color='black')

# Formatting
ax.set_title("Avg Cost Histogram with Count Labels", fontsize=14)
ax.set_xlabel("Avg Cost (₹)", fontsize=12)
ax.set_ylabel("Number of Restaurants", fontsize=12)
ax.set_xticks(bin_edges)

plt.tight_layout()
plt.show()
```
