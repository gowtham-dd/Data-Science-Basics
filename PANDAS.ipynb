```python
#extract particular element
import pandas as pd
#position based index
s1=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
print(s1[1])
#label based
print(s1['a'])
print(s1[::-1])
print(s1[0:3])
print(s1[:4])
print(s1['a':'d'])#last included in label
print(s1[['a','c','e']])#extract particular use list in order o do it
```

```python
#mathmatical operations
#missing data
import pandas as pd
x=pd.Series([1,2,3,4,6,8,9,0])
print(x)
x.isnull().sum()#count of null rec
```

```python
import pandas as pd
x.fillna(2.5)#replaces missing with 2.5
x.fillna(x.mean())
```

```python
#ffill() bfill()
x.ffill()#fill the null with previous record
x.bfill()#fill the null with backward value or below value
x.dropna()#drop the null value
```
