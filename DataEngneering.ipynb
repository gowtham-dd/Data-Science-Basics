What is PySpark? ðŸ”¥ðŸ
PySpark is the Python API for Apache Spark, an open-source distributed computing system. It enables big data processing and machine learning on massive datasets across multiple nodes using parallel computing.

What is RDD (Resilient Distributed Dataset)? âš¡
RDD is the fundamental data structure in Spark. It is an immutable, distributed collection of objects that can be processed in parallel.

```python
a=pd.Series([1,7,3,4,5],index=['a','b','c','d','e'])
print(a)
print(a.sum())
print(a.mean())
print(a.min())
print(a.max())
print(a.std())#standard deviation
print("***************************")
print(a.describe())
print(a.unique())
print(a.value_counts())
print(a.sort_values())
```

```python
b=pd.Series([5,4,6,2,7,1,7],index=['a','b','c','d','e','f','g'])
print(b.cumsum())
print("#################")#5,5+4,5+4+6,5+4+6+2.......
print(b.cummax())#5>4,5>6,6>2,6>7,7>1,7>7
print("#################")
print(b.cummin())
print("#################")
print(b.cumprod())#5,5*4,20*
```
