## Relational and Sub plots



### Difference Between Relational Plot and Subplot in Seaborn

**Relational Plot** and **Subplot** serve different purposes in data visualization using Seaborn and Matplotlib.

**Relational Plot** is a high-level plotting function in Seaborn used to display relationships between two or more variables. It is especially useful for visualizing trends and patterns in datasets. Common types of relational plots include scatter plots and line plots. Seabornâ€™s `relplot()` function allows for automatic creation of multiple related plots (facets) based on categorical variables. This makes it ideal for comparing how relationships vary across subgroups of data, such as comparing trends by gender or time.

On the other hand, a **Subplot** refers to the manual creation of multiple plot panels within a single figure using Matplotlib. It gives the user full control over the layout and appearance of each plot. Subplots are used when there is a need to display multiple different plots together in a customized grid arrangement, even if the plots are unrelated.

In summary, **relational plots** are data-driven and automated for exploring relationships across variables, while **subplots** are layout-driven and manually arranged to organize multiple visualizations together.


```python
## relational plot explain the relation between variables
#relplot()
# add a time row with it and show diff

sns.relplot(x=data.total_bill,y=data.tip,hue=data.smoker,col=data.sex,row=data.time)
plt.show()
```

```python
## subplot uses extra column and row
sns.relplot(x=data.total_bill,y=data.tip,hue=data.smoker,col=data.smoker,row=data.sex,kind='scatter')
plt.show()
```

```python
## subplot uses extra column and row
sns.relplot(x=data.total_bill,y=data.tip,hue=data.smoker,col=data.smoker,row=data.day,kind='scatter')
plt.show()
```

## wrapping and ordering

```python
sns.relplot(x=data.total_bill,y=data.tip,hue=data.smoker,col=data.day,kind='scatter',col_wrap=4,col_order=['Sun','Sat','Fri','Thur'])
plt.show()
##col_warp
##col_order and row cant be used at a time
```

## Size

```python
## subplot uses extra column and row
#size give dark color to large used data use alpha to transperency
sns.relplot(data=data, x='total_bill', y='tip', col='smoker', row='sex', kind='scatter', size='size', hue='size')
plt.show()
```

```python
sns.relplot(
    data=data,
    x='total_bill',
    y='tip',
    col='smoker',
    row='sex',
    kind='scatter',
    hue='size',
    size='size',
    alpha=0.6
)
plt.show()
```

## Lines

```python
# just change kind to line
sns.relplot(
    data=data,
    x='total_bill',
    y='tip',
    hue='day',
    markers=True,
    kind='line',
    dashes=False

)
plt.show()
```

```python
sns.relplot(
    data=data,
    x='day',
    y='total_bill',
    kind='scatter'
)

plt.show()

sns.relplot(
    data=data,
    x='day',
    y='total_bill',
    kind='line'
)
# gives default mean region that is shaded
#low standard Deviation best decision making
plt.show()
sns.relplot(
    data=data,
    x='day',
    y='total_bill',
    kind='line',
    ci=None
)

plt.show()
```

## Cat Plot

```python
sns.catplot(data=data,x='day',y='total_bill',kind='box')
plt.show()

sns.catplot(data=data,x='sex',kind='count')
plt.show()

sns.catplot(data=data,x='day',y='total_bill',kind='bar')
plt.show()

sns.catplot(data=data,x='day',y='total_bill',kind='violin')
plt.show()

sns.catplot(data=data,x='day',y='total_bill',kind='violin',inner='quartile')
plt.show()

sns.catplot(data=data,x='day',y='total_bill',kind='swarm')
plt.show()

```

## Bar Plot

```python
#1 category needed
sns.barplot(data=data,x='day',y='total_bill')
plt.show()

sns.barplot(data=data,x='day',y='total_bill',ci=None)
plt.show()

#to change vertical to horizontal change x and y

sns.barplot(data=data,y='day',x='total_bill',ci=None)
plt.show()
```

## Box

```python
sns.catplot (data=data,x='day',y='total_bill',kind='box')
plt.show()

#whis aks whishkers that is the box in middle can be changed
# default is 1.5 u can extend upto 2.0 whis=2.0
# showing percentile whis=[5,95]

sns.catplot(data=data,x='day',y='total_bill',kind='box',whis=[5,95])
plt.show()


sns.catplot(data=data,x='day',y='total_bill',kind='box',whis=2.0)
plt.show()
```

## Point Plot
