## Data Reshaping

```python
#  Pivot to create new df
# Melt reverse to original form
# stack
# unstack both used in multi level indexing
```

```python
#import
import pandas as pd
```

```python
data={

      'Date' :['2025-01-01', '2025-01-01', '2025-01-02', '2025-01-02'],
    'Category': ['A', 'B', 'C', 'B'],
    'Value': [10,20,30,25]
}
df=pd.DataFrame(data)
df
```

## Data Reshaping

## Data Sub-Setting and Indexing

```python
subset=df[df['Value']>20]
subset
```

## Pivot and melting

```python
# pivot is a way to reorganize the data from long to wide format
# long means more rows wide is more columns
```

```python
pivot_df=df.pivot_table(index='Date', columns='Category',values='Value')
# pivot_df
pivot_df.index
pivot_df.columns
```

## Melting

```python
# reorganie data from wide to long
# 1 reset index before melting
pivot_df.reset_index(inplace=True)
pivot_df.columns_name=None
```

```python
melted_df=pivot_df.melt(id_vars='Date',var_name='Category',value_name='Value')
melted_df
```

## Stacking & Unstacking

```python
# stackin used in multi level index
#eg in pivot table there is more index column
```

```python
df.set_index(['Date','Category'])
```

```python
stacked_data=df.set_index(['Date','Category']).stack()
stacked_data
```

```python
stacked_data['2025-01-01']
```

```python
un_stack_df=stacked_data.unstack()
un_stack_df
```

```python
un_stack_df.reset_index()
```

## Data Aggregation

```python
data={

      'Category':['Electronics','Electronics','Clothing','Clothing','Electronics'],
    'Product':['Laptop','Phone','Shirt','Shirt','Tablet'],
    'Sales':[1000,1500,200,300,800],
     'Profit':[300,1000,50,100,300]
}
df=pd.DataFrame(data)
df
```

```python
#Grouping data usong group by
grouped_data=df.groupby('Category')
grouped_data.sum()
```

```python
## Applying Aggregation to grouped data
grouped_data['Sales'].max()
```

```python
df.groupby('Category')['Profit'].max()
total_sales=grouped_data['Sales'].sum()
total_sales
```

```python
avg_sales =grouped_data['Sales'].mean()
avg_sales
```

```python
Count_prod =grouped_data['Category'].count()
Count_prod
```

```python
grouped_data.get_group('Clothing')
grouped_data.describe()
```

```python
# Merge is done by 4 types of joins same as sql
authors_data={

              'Author_ID':[1,2,3,4],
    'Author_Name':['John Doe','Jane Smith','Alice Johnson','Bob Brown'],'BookTitle' :['A','b','C','D']
              }
books_data={
}
author=pd.DataFrame(authors_data)
```

```python
books_data={

              'Book_ID':[1,2,3,4],
    'Book_Title':['Book A','Book B','Book C','Book D'],
    'Author_ID':[1,2,3,4],
              'Nationality':['','','','']}
books=pd.DataFrame(books_data)
```

```python
print(author,'\n\n',books)
```

```python
#default inner join that is merge all

default_merge=author.merge(books)
default_merge
```

```python

merge1=author.merge(books,on='Author_ID')
merge1
```

```python

Custommerge=author.merge(books,on='Author_ID',how='inner',suffixes=('_1','_2'))
Custommerge
```

```python

Outermerge=author.merge(books,on='Author_ID',how='outer',suffixes=('_1','_2'))
Outermerge
```

```python

Leftmerge=author.merge(books,on='Author_ID',how='left',suffixes=('_1','_2'))
Leftmerge
```

```python

Rightmerge=author.merge(books,on='Author_ID',how='right',suffixes=('_1','_2'))
Rightmerge
```

## Concat

```python
#add matching columns by default it uses outer join

concat_df=pd.concat([author,books],axis=0)
concat_df
```

```python
concat_df=pd.concat([author,books],join='inner')#intersection
concat_df
#only matching columns
```

```python
books.drop('Nationality',inplace=True,axis=1)
```

```python
books
```

```python
concat_df=pd.concat([author,books],ignore_index=True)#intersection
concat_df
```
