```python
# prompt: generate a bar plot for top 20 city with ratings

import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
data = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/refs/heads/main/zomato/zomato.csv")

# Assuming 'Aggregate rating' represents ratings
# Group by city and calculate the average rating
city_ratings = data.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)

# Select top 20 cities
top_20_cities = city_ratings[:20]

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_20_cities.index, top_20_cities.values, color='skyblue')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Top 20 Cities with Highest Average Restaurant Ratings')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
